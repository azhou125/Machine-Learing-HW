```python
import csv


# Solution 1:
begin = 0
end = 4
dataSet=[]
labels=[]
with open('C:/Users/AngZhou/Desktop/Cornell_Course/AML/HW/HW0/Iris_flowers_dataset/iris.data.txt', 'rb') as f:
    for row in csv.reader(f):
        # print row
        if row != []:
            dataSet.append(map(float,row[begin:end]))
            labels.append(row[end])
print dataSet
print labels

# Solution 2:
# dataSet=[]
# labees=[]
# with open('C:\Users\AngZhou\Desktop\Cornell_Course\AML\HW\Iris_flowers_dataset\iris.data.txt', 'rb') as f:
    # for line in f:
        # array=line.rstrip().split(',')
        # if len(array) == 5:
            # dataSet.append(array[0:4])
            # labels.append(array[4])
# print dataSet
# print labees
```

```python
%matplotlib inline

from matplotlib import pyplot as plt 
import numpy as np
npDataSet = np.array(dataSet)
colors=[]
# assuming the obs haven't been sorted:
for i,rec in enumerate(npDataSet):
    if labels[i]=='Iris-setosa':
        colors.append('r')
    elif labels[i]=='Iris-versicolor':
        colors.append('g')
    else:
        colors.append('b')

fields_vec=[]
for i in range(len(npDataSet[0])):
    fields_vec.append(npDataSet[:,i])

plt.figure(1,figsize=(18,18))
plt.figure()
plt.figure(1,figsize=(1,1))
label_names=['Sepal length in cm','Sepal width in cm','Petal length in cm','Petal width in cm']
for a in range(4):
    for b in range(4):
        if a!=b:
            plt.subplot(4,4,a*4+(b+1))
            plt.scatter(fields_vec[b], fields_vec[a], c=colors)
            plt.xlabel(label_names[b])
            plt.ylabel(label_names[a])
        else:
            plt.subplot(4,4,a*4+(b+1))
            plt.text(0.08, 0.48, label_names[a], dict(size=20))

plt.savefig("plot.png")
```
