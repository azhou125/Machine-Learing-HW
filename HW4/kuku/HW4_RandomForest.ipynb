```python
import numpy as np
from PIL import Image
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
import random
import itertools
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
image = mpimg.imread('MonaLisa.jpg') 
plt.imshow(image)
print image.shape
```

```python
### (b)
sample_num = 5000
row = image.shape[0]
column = image.shape[1]
points = list(itertools.product(range(0, row), range(0, column)))
print(len(points))

train_set = random.sample(points, sample_num)
print train_set[0]
```

```python
### (c)
pixel_value = []
for i in range(sample_num):
    pixel_value.append(image[train_set[i]])
    pixel = np.array(pixel_value)

print pixel.shape
print pixel
```

```python
# Processing and Mapping R, G, B at the same time
sampleDots = np.zeros([row, column, 3], dtype = np.uint8)
sampleDots.fill(255)
for i in range(sample_num):
    sampleDots[train_set[i][0]][train_set[i][1]] = pixel[i]
    
plt.imshow(sampleDots)
plt.title("Training Data Points")
plt.show()
```

```python
pixel_rescale = pixel/255.0
print pixel_rescale
```

```python
def random_forest(depth, num_tree):
    clf = RandomForestRegressor(random_state=0, n_estimators=num_tree, max_depth = depth)
    clf.fit(train_set, pixel)
    pred = clf.predict(points) / 255
    pred = pred.reshape(row, column, 3)
    return pred
```

```python
##(d)
pred_max_depth = random_forest(None, 1)
plt.title("Final Image")
plt.imshow(pred_max_depth)
plt.show()
```

```python
##(e)(i)
depth_list = [1, 2, 3, 5, 10, 15]
plt.figure(figsize=[10, 10])
for i, depth in enumerate(depth_list):
    pred = random_forest(depth, 1)
    plt.subplot(2, 3, i + 1)
    plt.title("Single Tree/Depth " + str(depth))
    plt.imshow(pred)
plt.tight_layout()
```

```python
### (e)(ii)
tree_list = [1, 3, 5, 10, 100]
plt.figure(figsize=[10, 10])
for i, tree in enumerate(tree_list):
    pred = random_forest(7, tree)
    plt.subplot(2, 3, i + 1)
    plt.title("Depth 7/Tree " + str(tree))
    plt.imshow(pred)
plt.tight_layout()
```

```python
### (e)(iii)
knn = KNeighborsRegressor(n_neighbors=1)
knn.fit(train_set, pixel)
pred_knn = knn.predict(points).reshape(row, column, 3) / 255
plt.imshow(pred_knn)
plt.title('1-NN Regressor')
```

```python
def random_forest_pruning(split, leaf):
    clf = RandomForestRegressor(random_state=0, n_estimators=50, max_depth = 25, min_samples_split=split, min_samples_leaf=leaf)
    clf.fit(train_set, pixel)
    pred = clf.predict(points) / 255
    pred = pred.reshape(row, column, 3)
    return pred
```

```python
### (e)(iiii)
split_list = [5, 15, 30]

plt.figure(figsize=[10, 10])
for i, split in enumerate(split_list):
    pred = random_forest_pruning(split, 1)
    plt.subplot(1, 3, i + 1)
    plt.title("Min samples split : " + str(split))
    plt.imshow(pred)
plt.tight_layout()
```

```python
leaf_list = [1, 5, 10]

plt.figure(figsize=[10, 10])
for i, leaf in enumerate(leaf_list):
    pred = random_forest_pruning(2, leaf)
    plt.subplot(1, 3, i + 1)
    plt.title("Min samples leaf : " + str(leaf))
    plt.imshow(pred)
plt.tight_layout()
```
